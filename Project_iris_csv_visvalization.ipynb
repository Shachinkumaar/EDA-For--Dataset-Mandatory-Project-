```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pyplot
import seaborn as sns
```

```python
df=pd.read_csv('iris.csv')
```

```python
df
```

```python
sns.countplot(x='species',data=df)
```

```python
df.plot(kind='scatter',x='sepal_length',y='sepal_width')
```

```python
sns.countplot(x='',hue='species',data=df,palette='RdBu_r')
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(df, hue="species", size=4) \
   .map(pyplot.scatter, "sepal_length", "sepal_width") \
   .add_legend();
pyplot.show();
```

```python
pyplot.close();
sns.set_style("whitegrid");
sns.pairplot(df, hue="species", size=5);
pyplot.show()
```

```python
#1D scatter plot
import numpy as np
df_setosa = df.loc[df["species"] == "setosa"];
df_virginica = df.loc[df["species"] == "virginica"];
df_versicolor = df.loc[df["species"] == "versicolor"];
#print(df_setosa["petal_length"])
pyplot.plot(df_setosa["petal_length"], np.zeros_like(df_setosa['petal_length']), 'o')
pyplot.plot(df_versicolor["petal_length"], np.zeros_like(df_versicolor['petal_length']), 'o')
pyplot.plot(df_virginica["petal_length"], np.zeros_like(df_virginica['petal_length']), 'o')

pyplot.show()
```

```python
#distribution based on petal length,sl,pw,sw
sns.FacetGrid(df, hue="species", size=5) \
   .map(sns.distplot, "petal_length") \
   .add_legend();
pyplot.show();
```

```python
sns.FacetGrid(df, hue="species", size=5) \
   .map(sns.distplot, "sepal_length") \
   .add_legend();
pyplot.show();
```

```python
sns.FacetGrid(df, hue="species", size=5) \
   .map(sns.distplot, "sepal_width") \
   .add_legend();
pyplot.show();
```

```python
sns.FacetGrid(df, hue="species", size=5) \
   .map(sns.distplot, "petal_width") \
   .add_legend();
pyplot.show();
```
